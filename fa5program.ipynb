```python
#function
def total(a, b):
  return a * b

#main
price = int(input("How much is your item: "))
quantity = int(input("How many of that item do you have: "))

total_value = total(price, quantity)
print("Your item costs", price, "and its quantity is", quantity, "so your total is", total_value )
```
